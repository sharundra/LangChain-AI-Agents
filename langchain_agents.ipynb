```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langchain_community.tools import DuckDuckGoSearchRun
import requests
```

```python
search_tool = DuckDuckGoSearchRun()
print(search_tool.invoke("Best films of 2025"))
```

```python
llm = ChatOpenAI()
print(llm.invoke("best films of 2025").content)
```

```python
from langchain_classic.agents import create_react_agent, AgentExecutor
from langchain_classic import hub
```

```python
# pull the standard react agent prompt from Langchain Hub. React(reasoning and action) is one the most famous desingn pattern to make agents.
prompt = hub.pull('hwchase17/react')
```

```python
#create the react agent manually with the prompt pulled earlier
agent = create_react_agent(
    llm = llm,
    tools = [search_tool],
    prompt = prompt
)
```

```python
# Wrap it with AgentExecutor
agent_executor = AgentExecutor(
    agent = agent,
    tools = [search_tool],
    verbose = True
)
```

```python
response = agent_executor.invoke({"input": "Best films of 2025"})
print(response)
print(response['output'])
```

```python
@tool
def get_weather_data(location: str) -> float:
    """
    Finds the weather data for a given location
    """
    url = f'https://api.weatherstack.com/current?access_key=f2baf0b886671a4c38920bb63564670e&query={location}'

    response = requests.get(url)
    return response.json()
```

```python
agent = create_react_agent(
    llm = llm,
    tools = [search_tool, get_weather_data],
    prompt = prompt
)
```

```python
agent_executor = AgentExecutor(
    agent = agent,
    tools = [search_tool, get_weather_data],
    verbose = True
)
```

```python
result = agent_executor.invoke({"input": "What is the most ancient city of the world and whagt is the weather there currently"})
print(result['output'])
```
